```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Create a sample DataFrame with at least 8 columns
# Replace this with your actual dataset if available
np.random.seed(2294984)  # Set a seed for reproducibility
data = pd.DataFrame(
    np.random.randint(1, 100, size=(100, 8)),  # 100 rows, 8 columns
    columns=["Column1", "Column2", "Column3", "Column4", "Column5", "Column6", "Column7", "Column8"]
)

# Step 2: Identify columns based on SID
col_x = "Column4"  # 4th column
col_y = "Column8"  # 8th column

# Step 3: Plot the bicolour features interaction diagram
plt.figure(figsize=(8, 6))
scatter = plt.scatter(
    data[col_x],
    data[col_y],
    c=data[col_x] + data[col_y],  # Colour by some interaction (e.g., sum of the two columns)
    cmap="coolwarm",  # Choose a colormap for bicolour effect
    edgecolor="k"
)
plt.colorbar(scatter, label="Interaction Value (Sum of Columns)")
plt.title(f"Bicolour Features Interaction Diagram: {col_x} vs {col_y}")
plt.xlabel(col_x)
plt.ylabel(col_y)
plt.grid(True)
plt.show()

# Lab Logbook: Display some sample data
print("Sample DataFrame:")
print(data.head())

print(f"\nBicolour Interaction Plot between '{col_x}' and '{col_y}' generated.")
```
